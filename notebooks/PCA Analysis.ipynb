# PCE Analysis

```python
import os
import sys

import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import seaborn as sns

sys.path.append("..")
import util_wsa_uncertainty

sns.set()
```

```python
os.chdir("..")
```

```python
knn_dataset = util_wsa_uncertainty.KnnUncertaintyDataset(
    input_map="AGONG", sat="ACE", real=0, daysahead=3
)
```

```python
pca = PCA(n_components=2)
data_2d = pca.fit_transform(knn_dataset.X)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(data_2d[:, 0], data_2d[:, 1], ".-", alpha=0.2)
plt.title("PCA Projection of k-NN State Space", fontweight="bold")
plt.xlabel("PCA Component 1 (km/s)")
plt.ylabel("PCA Component 2 (km/s)")
```
