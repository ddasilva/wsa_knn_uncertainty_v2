# Histogram of Predicted Uncertainties

This plots data written by the `process_uncertainties.py` script.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('../data/processed/processed_daysahead3_R000.csv')
df.head()
```

```python

bins = np.arange(35, 190, 2.5)

df_hist = pd.DataFrame()
df_hist['+6 Hours'] = df['sigma0']
df_hist['+1 Day'] = df['sigma3']
df_hist['+2 Days'] = df['sigma7']
df_hist['+3 Days'] = df['sigma11']


plt.figure(figsize=(8, 4))
ax = sns.histplot(df_hist, bins=bins, kde=True)
plt.xlabel(r'Predicted Uncertainty $\sigma$ (km/s)')
ax.get_legend().set_title('Time Since\nLast Observation')
plt.ylabel('Bin Count')
plt.title('Histogram of Predicted Uncertainties (2010-2020)', fontweight='bold')
plt.savefig('../plots/histogram_of_sigmas.png', dpi=300)
```
