# Histogram of Predicted Uncertainties

This plots data written by the `process_uncertainties.py` script.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/processed/processed_daysahead3_R000.csv')
```

```python
sns.set()

bins = np.arange(35, 190, 5)
sns.histplot(df.sigma, bins=bins, label='Input Map: ADAPT-GONG\nLocation: L1\nDays Ahead: 3\nRealization: 0', kde=True)
plt.legend()
plt.xlabel(r'Predicted Uncertainty $\sigma$ (km/s)')
plt.ylabel('Bin Count')
plt.title('Histogram of Predicted Uncertainties (2010-2020)', fontweight='bold')
```
