# Histogram of Predicted Uncertainties

This plots data written by the `prepare_histogram_plot_data.py` script.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/histogram_plot_data.csv')
```

```python
bins = np.arange(35, 190, 5)
sns.histplot(df.sigma, bins=bins, label='Input Map: ADAPT-GONG\nLocation: L1\nDays Ahead: 3', kde=True)
plt.legend()
plt.xlabel(r'Predicted Uncertainty $\sigma$ (km/s)')
plt.ylabel('Bin Count')
plt.title('Histogram of Predicted Uncertainties (2010-2020)', fontweight='bold')
```
